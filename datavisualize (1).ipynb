```python
# Import libraries as per requirement...
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Import the data in jupyter notebook
data = pd.read_csv(r"C:\Users\ADMIN\Downloads\mobile_sales.csv",encoding="latin1")
# Make a dataframe of the import data
df = pd.DataFrame(data)
# check the datasets informations
check_data_info=df.info()
print(check_data_info)
# Handle missing values and clean the data
# check if any null values in the columns
check_null=df.isnull().sum()
print(check_null) # their is no null values in the dataset
# Check duplicates
duplicates_=df.duplicated() 
print(duplicates_)# their is no duplicates in the dataset
# Group by gender and aggregate the total units sold 
gender_total_unit_sold=df.groupby("CustomerGender")["UnitsSold"].sum()
# Creating the pie chart
plt.figure(figsize=(8, 6)) 
plt.pie(gender_total_unit_sold,labels=gender_total_unit_sold.index,autopct='%1.1f%%', startangle=120, colors=['skyblue', 'lightgreen'])
plt.title('Total Units Sold by Customer Gender')
# Display the plot
plt.show()
# Make a line chart
payment_method_total_revenue=df.groupby("PaymentMethod")["TotalRevenue"].sum()
plt.figure(figsize=(10, 6)) 

plt.plot(payment_method_total_revenue.index, payment_method_total_revenue.values, marker='o', linestyle='-', color='red')

plt.title('Total Revenue by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45) 
plt.grid(True) 

# Display the plot
plt.show()

payment_method_sum_unit_sold=df.groupby("PaymentMethod")["UnitsSold"].sum()
plt.figure(figsize=(10, 6)) 

payment_method_sum_unit_sold.plot(kind='bar', color='skyblue')

plt.title('Total Units Sold by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Units Sold')
# Display the plot
plt.show()

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
# Extract Month Name
df['MonthName'] = df['Date'].dt.strftime('%B')
month_sum_of_units_solds=df.groupby("MonthName")["UnitsSold"].sum()
plt.figure(figsize=(5,8))
month_sum_of_units_solds.plot(kind="bar",color="yellow")
plt.title("Units Sold According To Months")
plt.xlabel("MonthNames")
plt.ylabel("UnitsSold")
plt.show()
# Top 8 brands Units sold
brand_units_sum=df.groupby("Brand")["UnitsSold"].sum()
top_8_brands=brand_units_sum.sort_values(ascending=False).head(8)
# Making Bar Chart of top 8 Brands
plt.figure(figsize=(5,8))
top_8_brands.plot(kind="bar",color="black")
plt.title("Top 8 Brands")
plt.xlabel("Brands")
plt.ylabel("UnitsSold")
plt.show()
# TOP 8 LOCATION WISE UNITS SOLD
location_unitssold=df.groupby("Location")["UnitsSold"].sum()
top_8_units_sold=location_unitssold.sort_values(ascending=False).head(8)
# Make scatter plot chart
plt.figure(figsize=(5,8))
top_8_units_sold.plot(kind="bar",color="orange")
plt.title("top 8 units sold by location")
plt.xlabel("Locations")
plt.ylabel("UnitsSold")
plt.show()
# Bottom 8 brands units sold
brandunitssold=df.groupby("Brand")["UnitsSold"].sum()
bottom_8_brands_unitssold=brandunitssold.sort_values(ascending=True).head(8)
plt.figure(figsize=(5,8))
bottom_8_brands_unitssold.plot(kind="bar",color="green")
plt.title("Bottom 8 brands unitssold")
plt.xlabel("Brands")
plt.ylabel("Unitssold")
plt.show()
# Bottom 5 location wise unitssold
locationunitssold=df.groupby("Location")["UnitsSold"].sum()
bottom_5_locationwise_unitssold=locationunitssold.sort_values(ascending=True).head(5)
plt.plot(bottom_5_locationwise_unitssold.index,bottom_5_locationwise_unitssold.values,marker="o",linestyle="-",color="purple")
plt.title("Bottom 5 location wise unitssold")
plt.xlabel("Locations")
plt.ylabel("Unitssolds")
plt.show()








```
